```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    data = pd.read_csv("post_pandemic_remote_work_health_impact_2025.csv")
except FileNotFoundError:
    print("Error: The file was not found. Please ensure the file exists in the current directory.")
    exit(1)

print(data.shape, data.columns)
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x="Mental_Health_Status", y="Social_Isolation_Score", data=data, palette="Blues_d", hue="Work_Arrangement", errorbar="sd")
plt.title("Impact of Remote Work on Mental Health and Social Isolation (2025)")
plt.xlabel("Mental Health Status")
plt.ylabel("Social Isolation Score")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("mental_health_work_correlation.png")
plt.show()
```

**Description:**

This bar chart illustrates the relationship between different mental health conditions and the average social isolation score across three work arrangements  Onsite, Hybrid, and Remote  in 2025 (Post-Pandemic data).

**X-axis**: Mental Health Status categories (Stress Disorder, ADHD, Burnout, Anxiety, PTSD, Depression).

**Y-axis**: Average Social Isolation Score (higher scores indicate greater perceived isolation).

**Bars**: Represent the mean social isolation score for each work arrangement within each mental health category.

**Error bars**: Show variability in the data (likely standard deviation or standard error).

**Key insight**:

Across most mental health conditions, remote workers tend to have **higher** average social isolation scores compared to onsite and hybrid workers.
Notably, individuals reporting **depression** exhibit the **highest** average social isolation scores across all work arrangements, with remote workers showing the most pronounced levels.

```python
print(data["Work_Life_Balance_Score"].min(), data["Work_Life_Balance_Score"].max())

sns.set_theme(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.barplot(x="Industry", y="Work_Life_Balance_Score", data=data, hue="Burnout_Level", palette="viridis", errorbar="sd")
plt.xlabel("Industry")
plt.ylabel("Work Life Balance Score")
plt.ylim(1, 5)
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("industry_to_worklifebalance_graph.png")
plt.show()
```

**Description:**

Work-Life Balance Across Industries by Burnout Level (2025)

The bar chart illustrates the **average Work-Life Balance Score (15 scale)** across various industries, segmented by burnout level (**High, Medium, Low**).  

- **X-axis:** Industry sectors (e.g., Professional Services, Education, Technology).  
- **Y-axis:** Work-Life Balance Score (1 = poor, 5 = excellent).  
- **Color coding (Viridis palette):**
  -  High Burnout
  -  Medium Burnout
  -  Low Burnout  

**Key Observations**
- Across most industries, the **Work-Life Balance Score remains clustered between 2.8  3.4**, regardless of burnout level.  
- Some variation is visible in **Customer Service and Marketing**, where employees with **Low Burnout** report slightly higher balance.  
- The error bars show standard deviation, suggesting a wide range of experiences within each industry.  

This visualization suggests that while burnout influences work-life balance, the **industry factor itself plays a consistent role across burnout groups**.

```python
from matplotlib.pyplot import viridis

physical_issues = ["Shoulder Pain", "Neck Pain", "Back Pain", "Shoulder Pain", "Eye Strain", "Wrist Pain"]
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x="Physical_Health_Issues", y="Age", data=data, order=physical_issues, hue="Physical_Health_Issues", legend=False, palette="plasma", errorbar="sd", estimator="mean")
plt.xlabel("Physical isssues")
plt.ylabel("Age")
plt.ylim(22, 66)
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("health_problems_age.png")
plt.show()
```

**Description:**

This bar chart shows the relationship between different physical issues and the average age of individuals reporting them.

**X-axis:** Physical issues (Shoulder Pain, Neck Pain, Back Pain, Eye Strain, Wrist Pain).

**Y-axis:** Age (average age of individuals experiencing each physical issue).

**Bars:** Represent the mean age for each physical issue.

**Error bars:** Indicate variability in the data.

**Key insight:**

Back pain and eye strain are reported by slightly older individuals on average compared to other physical issues.

Shoulder and wrist pain show similar mean ages, while neck pain appears slightly lower.
 
The wide error bars suggest considerable variation in the ages of individuals experiencing these conditions.
